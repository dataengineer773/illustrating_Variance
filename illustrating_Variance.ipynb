```python
from collections.abc import Callable
```

```python
def update(probe: Callable[[], float],
           display: Callable[[float], None]) -> None:
    temperature = probe()
    # imagine lots of control code here
    display(temperature)
```

```python
def probe_ok() -> int:
    return 77
```

```python
def display_wrong(temprature: complex) -> None:
    print(temprature)
```

```python
update(probe_ok, display_wrong) # type error
```

```python
def display_ok(temprature: complex) -> None:
    print(temprature)
```

```python
update(probe_ok, display_ok) # OK
```
